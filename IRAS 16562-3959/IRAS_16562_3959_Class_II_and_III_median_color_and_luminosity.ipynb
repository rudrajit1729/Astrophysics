# IRAS 16562-3959 Class II and Class III Median Color and Luminosity

```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Loading the data for Class II and Class III young stellar objects
J_mag_classII, H_mag_classII, Ks_mag_classII = np.loadtxt('X_ray_class_II.txt', unpack=True,usecols=[3,5,7])
J_mag_classIII, H_mag_classIII, Ks_mag_classIII = np.loadtxt('X_ray_class_III.txt', unpack=True,usecols=[3,5,7])

# Calculating the color H-Ks for each class
H_Ks_classII = H_mag_classII - Ks_mag_classII
H_Ks_classIII = H_mag_classIII - Ks_mag_classIII

# Calculating the median of the color H-Ks for each class
median_classII = np.nanmedian(H_Ks_classII)
median_classIII = np.nanmedian(H_Ks_classIII)
```

```python
# Plotting the results
plt.figure(1, figsize = (15,7))
ax1 = plt.subplot(121)
ax1.hist(H_Ks_classII[~np.isnan(H_Ks_classII)], bins=10, edgecolor='k')
ax1.axvline(median_classII, c='r')
ax1.text(0.65, 0.9, 'median = {0:.2f}'.format(median_classII),transform=ax1.transAxes, ha='right', va='top')
ax1.set_xlabel(r'H-K$_s$', fontsize=15)
ax1.set_ylabel('# X-Ray Sources', fontsize=15)
ax1.set_title('X-Ray Class II', fontsize=15)

ax2 = plt.subplot(122)
ax2.hist(H_Ks_classIII, bins=10, edgecolor='k')
ax2.axvline(median_classIII, c='r')
ax2.text(0.65, 0.9, 'median = {0:.2f}'.format(median_classIII),transform=ax2.transAxes, ha='right', va='top')
ax2.set_xlabel(r'H-K$_s$', fontsize=15)
ax2.set_ylabel('# X-Ray Sources', fontsize=15)
ax2.set_title('X-Ray Class III', fontsize=15)


plt.show()
```

```python
# Loading the data
lum_classII = np.loadtxt('X_ray_luminosity_class_II.txt', unpack=True,usecols=[1])
lum_classIII = np.loadtxt('X_ray_luminosity_class_III.txt', unpack=True,usecols=[1])

# Calculating the median of the X-ray luminosities for Class II and Class III young stellar objects
median_lum_classII = np.median(lum_classII)
median_lum_classIII = np.median(lum_classIII)
```

```python
# Plotting the results
plt.figure(2, figsize=(15,7))
ax1 = plt.subplot(121)
ax1.hist(lum_classII, bins=10, edgecolor='k')
ax1.axvline(median_lum_classII, c='r')
ax1.text(0.9, 0.9, 'median = {0:.2f}'.format(median_lum_classII), transform=ax1.transAxes, ha='right', va='top')
ax1.set_xlabel(r'Luminosity (erg.s$^{-1}$)', fontsize=15)
ax1.set_ylabel('# X-Ray Sources', fontsize=15)
ax1.set_title('X-Ray Class II', fontsize=15)


ax2 = plt.subplot(122)
ax2.hist(lum_classIII, bins=10, edgecolor='k')
ax2.axvline(median_lum_classIII, c='r')
ax2.text(0.9, 0.9, 'median = {0:.2f}'.format(median_lum_classIII), transform=ax2.transAxes, ha='right', va='top')
ax2.set_xlabel(r'Luminosity (erg.s$^{-1}$)', fontsize=15)
ax2.set_ylabel('# X-Ray Sources', fontsize=15)
ax2.set_title('X-Ray Class III', fontsize=15)


plt.show()
```
