At first, we use the NN2 algorithm to calculate the minimum lengths between each pair of sources that are part of a subcluster, and between each pair of sources that are not part of a subcluster.

```python
# Plotting the results of the NN2 algorithm
fig = plt.figure(1, figsize=(14,7))
fig.subplots_adjust(hspace=0.6, wspace=0.3)
ax1 = plt.subplot(121)
ax1.hist(distances_all_clusters[:,1], bins=15, edgecolor='k')
ax1.set_xlabel('NN2 Lengths (pc)', fontsize=15)
ax1.set_ylabel('# Sources', fontsize=15)
ax1.set_title('All Sources in Clusters', fontsize=15)

ax2 = plt.subplot(122)
ax2.hist(distances_all_not_cluster[:,1], bins=15, edgecolor='k')
ax2.set_xlabel('NN2 Lengths (pc)', fontsize=15)
ax2.set_ylabel('# Sources', fontsize=15)
ax2.set_title('All Sources Not in Clusters', fontsize=15)

plt.show()
```

We observe in the left panel, that for the sources associated with a subcluster, their NN2 lengths appear to have a mostly uniform distribution between 0 and 0.225 pc, with a peak at 0.05 pc. In the right panel, the NN2 lengths of the sources outside of a subcluster have a Poisson distribution with the mean at 0.15 pc. From these two histograms, we can deduce that the distance between the two closest sources that are part of a subcluster is in general smaller than for two sources outside of a subclus- ter, meaning that the surface density in subclusters is higher than outside of the subclusters.
