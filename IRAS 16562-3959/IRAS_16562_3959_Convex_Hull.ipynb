# IRAS 16562-3959 Convex Hull

I applied the Convex Hull algorithm to calculate the area of each subcluster defined in the IRAS 16562-3959 region.

```python
# Importing the libraries
import numpy as np
from scipy.spatial import ConvexHull
import matplotlib.pyplot as plt
```

```python
# Loading the data for Cluster A in order to visualize the data structure
cluster_A_data = np.loadtxt('Cluster_A_physical_coordinates.txt')

# Printing the sets of physical coordinates for the first 5 sources
print(cluster_A_data[:5])
```

```python
# Defining the string range
str = 'ABCDEFGHIJK'

# Defining the figure dimensions
fig = plt.figure(1, figsize=(15, 13))
fig.subplots_adjust(hspace=0.6, wspace=0.6)

# Looping through letters, and also using enumerate to index
for i, letter in enumerate(str):
    # Loading the data for each subcluster
    data = np.loadtxt('Cluster_' + letter + '_physical_coordinates.txt')
    # Applying the Convex Hull algorithm on the data
    hull = ConvexHull(data)
    # Printing the area for each subcluster
    print('Cluster ' + letter + ' area = %.3f pc-squared' % hull.area)

    # Plotting the Convex Hull for each subcluster
    i+=1
    ax = fig.add_subplot(3, 4, i)
    ax.scatter(data[:, 0], data[:,1], alpha=0.8, s=20)

    for simplex in hull.simplices:
        ax.plot(data[simplex, 0], data[simplex, 1], 'k-')

    ax.plot(data[hull.vertices,0], data[hull.vertices,1], 'r--', lw=2)
    ax.plot(data[hull.vertices[0],0], data[hull.vertices[0],1], 'ro')

    ax.set_title('Cluster ' + letter, fontsize=12)
    ax.set_xlabel('pc', fontsize=12)
    ax.set_ylabel('pc', fontsize=12)

plt.show()
```
